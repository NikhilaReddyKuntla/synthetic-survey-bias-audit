```python
import pandas as pd
import numpy as np
```

```python
pew = pd.read_csv("/Users/shravani/Downloads/W157_Oct24/ATP W157.csv", 
                low_memory=False)
print(f"Shape: {pew.shape}")
pew.head()
```

```python
pew['age_group'] = pew['F_AGECAT'].map({1:'18-29', 2:'30-49', 3:'50-64', 4:'65+'})
pew['race']      = pew['F_RACETHNMOD'].map({1:'White', 2:'Black', 3:'Hispanic', 4:'Asian', 5:'Other'})
pew['income']    = pew['F_INC_TIER2'].map({1:'Low', 2:'Middle', 3:'Upper'})
pew['gender']    = pew['F_GENDER'].map({1:'Male', 2:'Female', 3:'Other'})

pew_clean = pew[['age_group','race','income','gender',
                'JOBSAT_W157','JOBSEC_W157','AIWRKOPPO_W157',
                'AIWRKFEEL_c_W157','AIWRKHEARD_W157',
                'CHATWRK_W157','WEIGHT_W157']].replace(99, np.nan)

pew_clean = pew_clean.dropna(subset=['age_group','race','income'])
print(f"✅ Clean shape: {pew_clean.shape}")
```

```python
import os
os.makedirs("../data/processed", exist_ok=True)
pew_clean.to_csv("../data/processed/pew_clean.csv", index=False)
print("✅ Saved!")
```
